```python
#Finding min-samples or K count by Elbow method
from sklearn.cluster import KMeans
df2 = df[['Annual Income (k$)','Spending Score (1-100)']]
wc=[]
for i in range(1,11):
    B = KMeans(n_clusters=i,init='k-means++',random_state = 2)
    B.fit(df2)
    wc.append(B.inertia_)
plt.xlabel("no of clusters")
plt.ylabel("wcss")
plt.title("The Elbow method")
plt.plot(range(1,11),wc)
```

```python
#heatmap on the dataset
sns.heatmap(df.corr(),cbar=True,annot=True)
```

```python
#visualizing old-data
plt.xlabel("Annual Income (k$)")
plt.ylabel("Spending Score (1-100)")
plt.title("Before Clustering")
plt.scatter(df2.iloc[:,0],df2.iloc[:,1])
```

```python
#Getting the required columns
df = df.iloc[:,-2:]
df
```

```python
#implementing K-means
from sklearn.cluster import KMeans
model = KMeans(n_clusters=5)
label1 = model.fit_predict(df)
label1
```

```python
#printing the centroids
model.cluster_centers_
```

```python
#plot the data
colors=['red','green','blue','orange','violet']
plt.xlabel("Annual Income (k$)")
plt.ylabel("Spending Score (1-100)")
plt.title("Analyis of Spending Score with K-Means")
for i in range(len(colors)):
    plt.scatter(df.iloc[label1==i,0],df.iloc[label1==i,1],s=10,c=colors[i])
plt.legend(['A','B','C','D','E'])
```

```python
#accuracy of K-Means
silhouette_score(df,label1)
```

```python
#using clustering algorithm DBSCAN from sklearn
from sklearn.cluster import DBSCAN
model = DBSCAN(eps = 5,min_samples=5)
model
```

```python
#fitting and predict the model 
labels = model.fit_predict(df)
labels
```

```python
#finding unique labels
np.unique(labels)
```

```python
#Visualizing the data
B = list(np.unique(labels))
import matplotlib.pyplot as plt
colors=['black','blue','red','pink','green','orange']
plt.xlabel("Annual Income (k$)")
plt.ylabel("Spending Score")
plt.title("Analyis of Mall vs Customers Dataset with DBSCAN")
C = ['A','B','C','D','E','F']
for i in range(len(colors)):
    plt.scatter(df.iloc[labels == B[i],0],df.iloc[labels == B[i],1],s=10,c=colors[i],label=C[i])
plt.legend()
plt.show()
```

```python
#accuracy of DBSCAN algorithm 
silhouette_score(df,labels)
```

Conclusion:

1. K-Means Clustering is more efficient than DBSCAN algorithm in this case.
